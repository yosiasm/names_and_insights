```python
import pandas as pd
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/kodewilayah/permendagri-72-2019/main/dist/base.csv',names=['code','location'])
```

```python
df
```

```python
df = df[df['code'] != '13.02.02.1036']
```

```python
prov = {}
kabkota = {}
kec = {}
des = {}
new_df = []
for index,row in df.iterrows():
    code = row['code'].split('.')
    #prov
    if len(code) == 1:
        prov[row['code']] = row['location']
    #kabkota
    elif len(code) == 2:
        kabkota[row['code']] = row['location']
    #kec
    elif len(code) == 3:
        kec[row['code']] = row['location']
    #des
    elif len(code) == 4:
        des[row['code']] = row['location']

for id in des:
    code = id.split('.')
    prov_name = prov[code[0]]
    kabkota_name = kabkota['.'.join(code[0:2])]
    kec_name = kec['.'.join(code[0:3])]
    new_df.append(dict(
        code=id,
        provinsi=prov_name,
        kabkota=kabkota_name,
        kecamatan=kec_name,
        desa=des[id]
    ))

new_df = pd.DataFrame(new_df)
new_df
```

```python
new_df.to_csv('desa.csv',index=False)
```

```python
# kecamatan 
new_df['kec_code'] = new_df['code'].apply(lambda x: '.'.join((x.split('.')[0:3])))
new_df
```

```python
new_df.groupby(['kec_code','provinsi','kabkota','kecamatan']).count().reset_index().to_csv('kecamatan.csv',index=False)
```
