```python
nik_prov_code = {
    "01": "11", #Daerah Istimewa Aceh
    "02": "12", #Sumatera Utara
    "03": "13", #Sumatera Barat
    "04": "14", #Riau
    "05": "15", #Jambi
    "06": "16", #Sumatera Selatan
    "07": "17", #Bengkulu
    "08": "18", #Lampung
    "09": "31", #DKI Jakarta
    "10": "32", #Jawa Barat
    "11": "33", #Jawa Tengah
    "12": "34", #Daerah Istimewa Yogyakarta
    "13": "35", #Jawa Timur
    "14": "51", #Bali
    "15": "52", #Nusa Tenggara Barat
    "16": "53", #Nusa Tenggara Timur
    "17": "61", #Kalimantan Barat
    "18": "62", #Kalimantan Tengah
    "19": "63", #Kalimantan Selatan
    "20": "64", #Kalimantan Timur
    "21": "71", #Sulawesi Utara
    "22": "72", #Sulawesi Tengah
    "23": "73", #Sulawesi Selatan
    "24": "74", #Sulawesi Tenggara
    "25": "81", #Maluku
    "26": "91", #Irian Jaya
}
def find_location(nik):
    if int(nik[0:2])!= 0 and int(nik[10:12]) <= 90 and int(nik[0:2]) < 11 :
        print(nik)
        return '{}.{}.{}'.format(nik_prov_code[nik[0:2]],nik[2:4],nik[4:6])
    else:
        return '{}.{}.{}'.format(nik[0:2],nik[2:4],nik[4:6])


df['kec_id'] = df['nik'].apply(find_location)
df
```
