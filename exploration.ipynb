```python
from plotly.subplots import make_subplots

def horizontal_bar_labels(categories):
    subplots = make_subplots(
        rows=len(categories),
        cols=1,
        subplot_titles=[x["name"] for x in categories],
        shared_xaxes=True,
        print_grid=False,
        vertical_spacing=(0.45 / len(categories)),y_title='yes'
    )
    subplots['layout'].update(
        width=550,
        plot_bgcolor='#fff',
    )

    # add bars for the categories
    for k, x in enumerate(categories):
        subplots.add_trace(dict(
            type='bar',
            orientation='h',
            y=[x["name"]],
            x=[x["value"]],
            text=["{:,.0f}".format(x["value"])],
            hoverinfo='text',
            textposition='auto',
            marker=dict(
                color="#7030a0",
            ),
        ), k+1, 1)

    # update the layout
    subplots['layout'].update(
        showlegend=False,
    )
    for x in subplots["layout"]['annotations']:
        x['x'] = 0
        x['xanchor'] = 'left'
        x['align'] = 'left'
        x['font'] = dict(
            size=12,
        )

    # hide the axes
    for axis in subplots['layout']:
        if axis.startswith('yaxis') or axis.startswith('xaxis'):
            subplots['layout'][axis]['visible'] = False

    # update the margins and size
    subplots['layout']['margin'] = {
        'l': 0,
        'r': 0,
        't': 20,
        'b': 1,
    }
    height_calc = 45 * len(categories)
    height_calc = max([height_calc, 350])
    subplots['layout']['height'] = height_calc
    subplots['layout']['width'] = height_calc

    return subplots
```

```python
categories = [
    {"name": "Musée du Louvre, Paris", "value": 10200000},
    {"name": "National Museum of China, Beijing", "value": 8610092},
    {"name": "Metropolitan Museum of Art, New York City", "value": 6953927},
    {"name": "Vatican Museums, Vatican City", "value": 6756186},
    {"name": "Tate Modern, London", "value": 5868562},
    {"name": "British Museum, London", "value": 5820000},
    {"name": "National Gallery, London", "value": 5735831},
    {"name": "National Gallery of Art, Washington D.C.", "value": 4404212},
    {"name": "State Hermitage Museum, Saint Petersburg", "value": 4220000},
    {"name": "Victoria and Albert Museum, London", "value": 3967566},
]
horizontal_bar_labels(categories)
```

```python
def repeating_name(name:str):
    if type(name) == float:
        return
    names = name.split(' ')
    if len(names) < 2:
        return
    firstname = name.split(' ')[0]
    lastname = name.replace(firstname+' ','')
    
    key = None
    for itr in range(len(firstname)+1):
        chunk = firstname[itr:itr+3]
        # print(chunk)
        if lastname.count(chunk) > 0 and len(chunk) > 2:
            # print(chunk, lastname)
            key = chunk
    return key
    
# repeating_name('FATMAWATI SRI NINGSIH')
repeating_name('wawan darmawan')
```

```python
names = con.sql('SELECT nama FROM entity_with_coordinate2').fetchdf()
names
```

```python
names['unique'] = names['nama'].apply(repeating_name)
names.value_counts('unique').head()
```

```python
def remove_gelar(name,index):
    if type(name) == float:
        return None
    name = ' '.join([x for x in name.split(' ') if '.' not in x])
    names = name.split(' ')
    if len(names) > index:
        return name.split(' ')[-1 - index]
    
names['lastname1'] = names['nama'].apply(remove_gelar, index=0)
names['lastname2'] = names['nama'].apply(remove_gelar, index=1)
names['lastname3'] = names['nama'].apply(remove_gelar, index=2)
names['lastname4'] = names['nama'].apply(remove_gelar, index=3)
names['lastname5'] = names['nama'].apply(remove_gelar, index=4)
```

```python
names.to_csv('names.csv',index=False)
```

```python
name_nodes = {}
def create_node(row,key1,key2):
    return dict(
                source=row[key1],
                target=row[key2],
                value=1
            )
def name_nodes_append(name_nodes,row,key1,key2):
    if row[key1] and row[key2]:
        key = row[key1]+'_'+row[key2]
        if key not in name_nodes:
            name_nodes[key] = create_node(row,key1,key2)
        else:
            name_nodes[key]['value'] += 1

    return name_nodes

for id,row in names[["1stname","2ndname","3rdname","4thname","5thname"]].iterrows():
    name_nodes = name_nodes_append(name_nodes,row,'1stname','2ndname')
    name_nodes = name_nodes_append(name_nodes,row,'2ndname','3rdname')
    name_nodes = name_nodes_append(name_nodes,row,'3rdname','4thname')
    name_nodes = name_nodes_append(name_nodes,row,'4thname','5thname')

    
name_nodes_df = pd.DataFrame([name_nodes[key] for key in name_nodes])
name_nodes_df
```

```python
name_nodes_df.to_csv('name_composition.csv',index=False)
```

```python
name_nodes_df_filtered = name_nodes_df[name_nodes_df['value']>=20]
name_nodes_df_filtered
```
